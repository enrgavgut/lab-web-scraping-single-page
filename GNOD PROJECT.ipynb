### 1. LAB WEB SCRAPPING SINGLE PAGE

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 1. Find URL and store it as a variable

url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 2. download html with a get request
response = requests.get(url)
# 2.1 chech if succesful
response.status_code # 200 status code means OK!
```

```python
#response.content
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4.2. check that the html code looks like it should
#soup
```

```python
# 5. retrieve/extract the desired info 

soup.select("cite.title") # In the webpage if we place the mouse on the tile we see that is cite.title
```

```python
soup.select("em.artist")
```

```python
# now we are going to get the song names 

#initialize empty lists
song = []

# define the number of iterations of our for loop
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("cite.title"))

songlist = soup.select("cite.title")

# # iterate through the result set and retrive all the data
for i in range(num_iter):
    song.append(songlist[i].get_text())

print(song)
```

```python

# same process to get artist names

#initialize empty lists
artist = []

# define the number of iterations of our for loop
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("em.artist"))

artistlist = soup.select("em.artist")

# # iterate through the result set and retrive all the data
for i in range(num_iter):
    artist.append(artistlist[i].get_text())

print(artist)
```

```python
# each list becomes a column # from a series of list we can create a dataframe pretty easily
chart = pd.DataFrame({"song":song, "artist":artist})
chart.head()
```

```python
chart.to_csv('chart')
```

##### 2.4 Build a function that gets the names and artists of all songs in a playlist

```python


def get_playlist_tracks(playlist_id):
    playlist = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = playlist['items'] # list of dictionaries including information about tracks
    while playlist['next']!=None: # whilere there is still a next link to go to
        playlist = sp.next(playlist) # we overwrite results with content of the next page
        tracks = tracks + playlist['items'] # we are not appending but putting together complete lists
    
    return tracks
```

##### 2.6 Create a DataFrame with name of song, artist and uri

### 3. LAB UNSUPERVISED LEARNING INTRO - CLUSTERING

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from matplotlib.lines import Line2D
```

```python
data_num = data_concat.drop(['song','artist','id','uri'], axis=1)
```

```python
data_num
```

```python
data_num.columns
```

```python
data_cat=data_concat.drop(['danceability', 'energy', 'loudness', 'speechiness', 'acousticness','instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms'], axis=1)
```

##### 3.1 Standarize numericals

```python
# The scale of "proline" is much higher than the scale of many other variables!
# K-Means is a distance based algorithm: we need to scale / normalize:

from sklearn.preprocessing import StandardScaler
X_prep = StandardScaler().fit_transform(data_num)
# We don´t do a train test split as we dont have any data to train with
# Now, all features will have the same weight.
pd.DataFrame(X_prep,columns=data_num.columns).head()
```

##### 3.2 Choose the best number of cluster

```python
from sklearn.cluster import KMeans
```

```python

# We are going to calculate differents inertia over same data and plote 
import numpy as np
K = range(2, 10)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k') # cambio la pendiente en valor de 3
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 10)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

##### 3.3 Clustering - 8 Clusters

```python
kmeans = KMeans(n_clusters=8, random_state=1234)
kmeans.fit(X_prep)
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(X_prep)

# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
plt.figure(figsize=(5, 5)) # Here we are ploting 2 features 
plt.scatter(X_prep[:, 0], X_prep[:, 5], c=clusters,alpha=0.5)
plt.show()
```

```python
kmeans.inertia_
```

##### 3.4 Assigning clusters to tracks

```python
data_num_cluster = data_num.assign(clusters=clusters)
```

```python
data_num_cluster
```

##### 3.5 Concat final data

```python
data_final = pd.concat([data_cat,data_num_cluster],axis=1)
```
