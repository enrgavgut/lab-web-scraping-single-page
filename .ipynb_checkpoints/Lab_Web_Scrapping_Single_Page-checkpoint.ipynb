##### 1. LAB WEB SCRAPPING SINGLE PAGE

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# 1. Find URL and store it as a variable

url = "https://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
# 2. download html with a get request
response = requests.get(url)
# 2.1 chech if succesful
response.status_code # 200 status code means OK!
```

```python
#response.content
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4.2. check that the html code looks like it should
#soup
```

```python
# 5. retrieve/extract the desired info (here, you'll paste the "Selector" you copied before to get the element that belongs to the top movie)
soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper")
```

```python
soup.select("cite.title") # In the webpage if we place the mouse on the tile we see that is cite.title
```

```python
soup.select("em.artist")
```

```python
#initialize empty lists
song = []

# define the number of iterations of our for loop
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("cite.title"))

songlist = soup.select("cite.title")

# # iterate through the result set and retrive all the data
for i in range(num_iter):
    song.append(songlist[i].get_text())

print(song)
```

```python
#initialize empty lists
artist = []

# define the number of iterations of our for loop
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("em.artist"))

artistlist = soup.select("em.artist")

# # iterate through the result set and retrive all the data
for i in range(num_iter):
    artist.append(artistlist[i].get_text())

print(artist)
```

```python
# each list becomes a column # from a series of list we can create a dataframe pretty easily
chart = pd.DataFrame({"song":song, "artist":artist})
chart.head()
```

```python
# # We need to build a prototype:
# 1) User inputs a song
# 2) Is it in the list?
#     Yes? Recommend another song from the list
#     No? No recommendations
```

```python
import random

hot_songs = chart['song'].tolist() # We need to convert chart['song'] into a list to interact with songname varible

song_name = input("Enter the name of the song you're looking for: ")

if song_name in hot_songs:
    print(song_name + ' ' + 'is in the list')
    print(" We recommend you listen to", random.choice(hot_songs))    
else:
    print(song_name + ' ' + 'is not in the list')
```
